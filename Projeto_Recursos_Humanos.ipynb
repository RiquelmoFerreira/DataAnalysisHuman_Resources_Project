## Analisando os Anos de Experiência x Salário

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

### Dataset

```python
Data_Set = pd.read_csv('Salary_Data.csv')
Data_Set.head()
```

```python
Data_Set.shape
```

### Valores Nulos

```python
Data_Set.isnull().sum()
```

### Estatísticas

```python
Data_Set.describe()
```

### Análise Salarial

```python
plt.figure(figsize=(10, 6))
plt.title('Salary Analysis')
sns.kdeplot(Data_Set['Salary'], color='green');
```

### Análise dos Anos de Experiência

```python
plt.figure(figsize=(10, 6))
plt.title('Years Experience Analysis')
sns.kdeplot(Data_Set['YearsExperience'],  color='blue');
```

### Boxplot

```python
plt.figure(figsize=(10, 5))
sns.boxplot(Data_Set['Salary'], orient='h')
plt.yticks([])
plt.xlabel('Salary');
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(Data_Set['YearsExperience'], orient='h')
plt.yticks([])
plt.xlabel('Years Experience');
```

### Scatterplot

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=Data_Set, x='Salary', y='YearsExperience');
```

```python
plt.figure(figsize=(10, 5))
sns.regplot(data=Data_Set, x='Salary', y='YearsExperience');
```

### Correlação de Pearson

```python
Correlation = Data_Set.corr()
plt.figure(figsize=(10, 5))
sns.heatmap(Correlation, annot=True, cbar=False);
```
